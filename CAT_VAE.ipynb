**CAT-VAE for Ultrasount Breast cancer dataset (Normal, bening, and malignant)**

```python
import os
import time
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import (Input, Dense, Lambda, Flatten, Reshape,
                                   Conv2D, Conv2DTranspose, Layer,
                                   MultiHeadAttention, LayerNormalization, Dropout,
                                   Concatenate)
from tensorflow.keras.models import Model
from tensorflow.keras import backend as K
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import matplotlib.pyplot as plt
from tensorflow.image import ssim, psnr
from tensorflow.keras.applications.inception_v3 import InceptionV3, preprocess_input
from scipy.linalg import sqrtm
from tensorflow.keras.callbacks import History
from tensorflow.keras.utils import plot_model

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Define image dimensions
img_shape = (128, 128, 3)
latent_dim = 100
max_images_per_class = 1000

# Performance tracking class
class PerformanceTracker:
    def __init__(self):
        self.metrics = {
            'training_time': [],
            'inference_time': [],
            'gpu_memory': [],
            'mse': [],
            'psnr': [],
            'ssim': [],
            'fid': []
        }

    def start_timer(self):
        self.start_time = time.time()

    def record_training_time(self):
        elapsed = time.time() - self.start_time
        self.metrics['training_time'].append(elapsed)
        return elapsed

    def record_inference_time(self, batch_size):
        elapsed = (time.time() - self.start_time) / batch_size
        self.metrics['inference_time'].append(elapsed)
        return elapsed

    def get_gpu_memory(self):
        try:
            from pynvml import nvmlInit, nvmlDeviceGetHandleByIndex, nvmlDeviceGetMemoryInfo
            nvmlInit()
            handle = nvmlDeviceGetHandleByIndex(0)
            info = nvmlDeviceGetMemoryInfo(handle)
            return info.used // 1024 // 1024  # Convert to MB
        except:
            return None

    def record_metrics(self, mse, psnr, ssim, fid):
        self.metrics['mse'].append(mse)
        self.metrics['psnr'].append(psnr)
        self.metrics['ssim'].append(ssim)
        self.metrics['fid'].append(fid)

    def print_summary(self, class_name):
        print(f"\n=== Performance Summary for {class_name.upper()} ===")
        print(f"Training time: {np.mean(self.metrics['training_time']):.2f} ± {np.std(self.metrics['training_time']):.2f} sec")
        print(f"Inference time: {np.mean(self.metrics['inference_time']) * 1000:.2f} ± {np.std(self.metrics['inference_time']) * 1000:.2f} ms per image")
        if any(self.metrics['gpu_memory']):
            print(f"GPU memory usage: {np.mean(self.metrics['gpu_memory']):.0f} ± {np.std(self.metrics['gpu_memory']):.0f} MB")
        print(f"Reconstruction quality:")
        print(f"- MSE: {np.mean(self.metrics['mse']):.4f}")
        print(f"- PSNR: {np.mean(self.metrics['psnr']):.2f} dB")
        print(f"- SSIM: {np.mean(self.metrics['ssim']):.4f}")
        print(f"- FID: {np.mean(self.metrics['fid']):.2f}")

# Function to Load Images
def load_images(directory, img_size):
    images = []
    for filename in os.listdir(directory):
        img_path = os.path.join(directory, filename)
        try:
            img = load_img(img_path, target_size=img_size)
            img = img_to_array(img) / 255.0
            images.append(img)
        except Exception as e:
            print(f"Error loading {img_path}: {e}")
    return np.array(images)

# Load datasets for all three classes
base_dir = '/content/drive/MyDrive/Breast'
class_dirs = {
    'normal': os.path.join(base_dir, 'normal'),
    'benign': os.path.join(base_dir, 'benign'),
    'malignant': os.path.join(base_dir, 'malignant')
}

# Check if directories exist and load images
x_train = {}
for class_name, data_dir in class_dirs.items():
    if not os.path.exists(data_dir):
        raise ValueError(f"Directory {data_dir} does not exist! Check the path.")
    x_train[class_name] = load_images(data_dir, (128, 128))
    print(f"Loaded {len(x_train[class_name])} images for class {class_name}")

# Custom Transformer Block Layer with Cross-Attention
class CrossAttentionTransformerBlock(Layer):
    def __init__(self, num_heads=8, key_dim=64, dropout_rate=0.1, **kwargs):
        super().__init__(**kwargs)
        self.num_heads = num_heads
        self.key_dim = key_dim
        self.dropout_rate = dropout_rate

        # Cross-attention layers
        self.cross_attn = MultiHeadAttention(num_heads=num_heads, key_dim=key_dim, dropout=dropout_rate)
        self.dropout1 = Dropout(dropout_rate)
        self.layernorm1 = LayerNormalization(epsilon=1e-6)

        # Feed-forward network
        self.ffn = tf.keras.Sequential([
            Dense(key_dim * 4, activation='relu'),
            Dense(key_dim),
            Dropout(dropout_rate)
        ])
        self.layernorm2 = LayerNormalization(epsilon=1e-6)

    def call(self, inputs):
        # Unpack inputs: latent_z is query, encoder_features are key/value
        latent_z, encoder_features = inputs

        # Cross-attention
        attn_output = self.cross_attn(
            query=latent_z,
            key=encoder_features,
            value=encoder_features
        )
        attn_output = self.dropout1(attn_output)
        out1 = self.layernorm1(latent_z + attn_output)

        # Feed-forward network
        ffn_output = self.ffn(out1)
        return self.layernorm2(out1 + ffn_output)

    def get_config(self):
        config = super().get_config()
        config.update({
            'num_heads': self.num_heads,
            'key_dim': self.key_dim,
            'dropout_rate': self.dropout_rate
        })
        return config

# Enhanced Encoder that returns intermediate features
def build_encoder():
    inputs = Input(shape=img_shape)

    # First conv block
    x = Conv2D(32, kernel_size=3, strides=2, padding='same', activation='relu')(inputs)

    # Second conv block (we'll use these features for cross-attention)
    encoder_features = Conv2D(64, kernel_size=3, strides=2, padding='same', activation='relu')(x)

    # Flatten and get latent distribution parameters
    x = Flatten()(encoder_features)
    x = Dense(256, activation='relu')(x)

    z_mean = Dense(latent_dim, name='z_mean')(x)
    z_log_var = Dense(latent_dim, name='z_log_var')(x)

    # Sampling function
    def sampling(args):
        z_mean, z_log_var = args
        batch = K.shape(z_mean)[0]
        dim = K.int_shape(z_mean)[1]
        epsilon = K.random_normal(shape=(batch, dim))
        return z_mean + K.exp(0.5 * z_log_var) * epsilon

    z = Lambda(sampling, output_shape=(latent_dim,), name='z')([z_mean, z_log_var])

    # Return both latent variables and intermediate features
    return Model(inputs, [z_mean, z_log_var, z, encoder_features], name='encoder')

# Enhanced Decoder with Cross-Attention Transformer
def build_decoder():
    # Two inputs: latent vector and encoder features
    latent_inputs = Input(shape=(latent_dim,))
    encoder_features_input = Input(shape=(32, 32, 64))  # Shape of encoder's intermediate features

    # Process latent vector
    x = Dense(16 * 16 * 64, activation='relu')(latent_inputs)
    x = Reshape((16, 16, 64))(x)

    # Prepare for cross-attention
    latent_flat = Reshape((16*16, 64))(x)  # Query (from latent)
    encoder_flat = Reshape((32*32, 64))(encoder_features_input)  # Key/Value (from encoder)

    # Apply cross-attention
    x = CrossAttentionTransformerBlock()([latent_flat, encoder_flat])
    x = Reshape((16, 16, 64))(x)

    # Upsampling
    x = Conv2DTranspose(64, kernel_size=3, strides=2, padding='same', activation='relu')(x)
    x = Conv2DTranspose(32, kernel_size=3, strides=2, padding='same', activation='relu')(x)
    outputs = Conv2DTranspose(3, kernel_size=3, strides=2, padding='same', activation='sigmoid')(x)

    return Model([latent_inputs, encoder_features_input], outputs, name='decoder_with_cat')

# Custom Loss Layer
class VAELossLayer(Layer):
    def __init__(self, **kwargs):
        super(VAELossLayer, self).__init__(**kwargs)

    def vae_loss(self, inputs, outputs, z_mean, z_log_var):
        reconstruction_loss = tf.keras.losses.binary_crossentropy(
            K.flatten(inputs), K.flatten(outputs))
        reconstruction_loss = tf.reduce_mean(reconstruction_loss) * np.prod(img_shape)

        kl_loss = -0.5 * tf.reduce_sum(
            1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var), axis=-1)
        return tf.reduce_mean(reconstruction_loss + kl_loss)

    def call(self, inputs):
        inputs, outputs, z_mean, z_log_var = inputs
        loss = self.vae_loss(inputs, outputs, z_mean, z_log_var)
        self.add_loss(loss)
        return outputs

def build_vae(encoder, decoder):
    inputs = Input(shape=img_shape)
    z_mean, z_log_var, z, encoder_features = encoder(inputs)
    outputs = decoder([z, encoder_features])
    outputs = VAELossLayer()([inputs, outputs, z_mean, z_log_var])
    return Model(inputs, outputs, name='vae_with_cat')

# Function to calculate evaluation metrics
def evaluate_model(encoder, decoder, vae, x_test):
    reconstructed_images = vae.predict(x_test)

    mse_scores = np.mean(np.square(x_test - reconstructed_images), axis=(1, 2, 3))
    psnr_scores = [psnr(x_test[i], reconstructed_images[i], max_val=1.0).numpy() for i in range(len(x_test))]
    ssim_scores = [ssim(x_test[i], reconstructed_images[i], max_val=1.0).numpy() for i in range(len(x_test))]

    avg_mse = np.mean(mse_scores)
    avg_psnr = np.mean(psnr_scores)
    avg_ssim = np.mean(ssim_scores)

    return avg_mse, avg_psnr, avg_ssim

# Load InceptionV3 model for FID calculation
fid_model = InceptionV3(include_top=False, pooling='avg', input_shape=(128, 128, 3))

# Function to compute FID Score
def calculate_fid(real_images, generated_images):
    # Preprocess images for InceptionV3
    real_images = preprocess_input(real_images)
    generated_images = preprocess_input(generated_images)

    # Compute feature representations
    act_real = fid_model.predict(real_images)
    act_generated = fid_model.predict(generated_images)

    # Calculate mean and covariance statistics
    mu_real, sigma_real = np.mean(act_real, axis=0), np.cov(act_real, rowvar=False)
    mu_generated, sigma_generated = np.mean(act_generated, axis=0), np.cov(act_generated, rowvar=False)

    # Compute squared difference of means
    mean_diff = np.sum((mu_real - mu_generated) ** 2)

    # Compute sqrt of product of covariances
    covmean, _ = sqrtm(sigma_real.dot(sigma_generated), disp=False)

    # Check for imaginary values
    if np.iscomplexobj(covmean):
        covmean = covmean.real

    # Compute FID score
    fid_score = mean_diff + np.trace(sigma_real + sigma_generated - 2 * covmean)
    return fid_score

# Train separate VAEs for each class
for class_name, images in x_train.items():
    tracker = PerformanceTracker()
    print(f"\n=== Training CAT-VAE for class: {class_name.upper()} ===")
    print(f"Training samples: {len(images)}")

    # Build models
    encoder = build_encoder()
    decoder = build_decoder()
    vae = build_vae(encoder, decoder)
    vae.compile(optimizer='adam')

    # Start training timer
    tracker.start_timer()

    # Train the model and store history
    history = vae.fit(images, epochs=500, batch_size=64, verbose=1)

    # Plot training history
    plt.figure(figsize=(10, 6))
    plt.plot(history.history['loss'], label='Training Loss')
    plt.title(f'Training Loss Curve for {class_name.upper()}')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    # Save the loss curve plot
    loss_curve_path = os.path.join(base_dir, f'loss_curve_{class_name}.png')
    plt.savefig(loss_curve_path)
    plt.close()
    print(f"Saved loss curve plot at: {loss_curve_path}")

    # Record training time
    training_time = tracker.record_training_time()
    print(f"Training completed in {training_time:.2f} seconds")

    # Record GPU memory
    gpu_mem = tracker.get_gpu_memory()
    if gpu_mem:
        tracker.metrics['gpu_memory'].append(gpu_mem)
        print(f"GPU memory used: {gpu_mem} MB")

    # Create a directory to save models for this class
    models_dir = os.path.join(base_dir, f'saved_models_VAE-CAT_Breast{class_name}')
    os.makedirs(models_dir, exist_ok=True)

    # Save models
    encoder.save(os.path.join(models_dir, 'encoder.h5'))
    decoder.save(os.path.join(models_dir, 'decoder.h5'))
    vae.save(os.path.join(models_dir, 'vae.h5'))

    # Save model architecture diagram
    architecture_path = os.path.join(models_dir, 'vae_architecture.png')
    try:
        plot_model(vae, to_file=architecture_path, show_shapes=True, show_layer_names=True)
        print(f"Saved model architecture diagram at: {architecture_path}")
    except Exception as e:
        print(f"Error saving model architecture diagram: {e}")

    print(f"Saved models for {class_name} in {models_dir}")

    # Create output directory for generated images
    output_dir = os.path.join(base_dir, f'generated_VAE-CAT_Breast{class_name}')
    os.makedirs(output_dir, exist_ok=True)

    # Calculate how many images to generate
    num_real_images = len(images)
    num_to_generate = max(0, max_images_per_class - num_real_images)

    if num_to_generate > 0:
        print(f"Generating {num_to_generate} images for class {class_name}")

        # Generate latent vectors
        latent_vectors = np.random.normal(size=(num_to_generate, latent_dim))

        # Need encoder features for generation - use average features from training set
        _, _, _, sample_encoder_features = encoder.predict(images[:min(100, len(images))])
        avg_encoder_features = np.mean(sample_encoder_features, axis=0, keepdims=True)
        repeated_features = np.repeat(avg_encoder_features, num_to_generate, axis=0)

        # Generate images
        generated_images = decoder.predict([latent_vectors, repeated_features])

        # Save generated images
        for i in range(num_to_generate):
            img_path = os.path.join(output_dir, f'generated_{i}.png')
            plt.imsave(img_path, generated_images[i])

    # Evaluate the model
    print(f"\nEvaluating model for class: {class_name}")
    x_test_sample = images[:min(10, len(images))]

    # Measure inference speed
    tracker.start_timer()
    _ = vae.predict(x_test_sample)
    inference_time = tracker.record_inference_time(len(x_test_sample))
    print(f"Inference speed: {inference_time*1000:.2f} ms per image")

    # Calculate quality metrics
    mse, psnr_score, ssim_score = evaluate_model(encoder, decoder, vae, x_test_sample)
    print(f"Reconstruction quality:")
    print(f"- MSE: {mse:.4f}")
    print(f"- PSNR: {psnr_score:.2f} dB")
    print(f"- SSIM: {ssim_score:.4f}")

    # Calculate FID score
    if len(x_test_sample) > 0:
        # Generate comparable number of images for FID
        latent_vectors = np.random.normal(size=(len(x_test_sample), latent_dim))
        _, _, _, sample_encoder_features = encoder.predict(x_test_sample)
        generated_for_fid = decoder.predict([latent_vectors, sample_encoder_features])
        fid_score = calculate_fid(x_test_sample, generated_for_fid)
        print(f"- FID: {fid_score:.2f}")
    else:
        fid_score = float('nan')

    # Record all metrics
    tracker.record_metrics(mse, psnr_score, ssim_score, fid_score)

    # Print summary
    tracker.print_summary(class_name)

# Create and save a comprehensive system diagram
try:
    # Create a temporary model that shows the full flow
    inputs = Input(shape=img_shape)
    z_mean, z_log_var, z, encoder_features = encoder(inputs)
    outputs = decoder([z, encoder_features])
    full_system = Model(inputs, outputs, name='full_vae_system')

    system_diagram_path = os.path.join(base_dir, 'full_system_architecture.png')
    plot_model(full_system, to_file=system_diagram_path, show_shapes=True, show_layer_names=True)
    print(f"\nSaved full system architecture diagram at: {system_diagram_path}")
except Exception as e:
    print(f"\nError creating full system diagram: {e}")

# Display all loss curves at the end
plt.figure(figsize=(15, 10))
for i, class_name in enumerate(x_train.keys()):
    loss_curve_path = os.path.join(base_dir, f'loss_curve_{class_name}.png')
    if os.path.exists(loss_curve_path):
        img = plt.imread(loss_curve_path)
        plt.subplot(2, 2, i+1)
        plt.imshow(img)
        plt.axis('off')
        plt.title(f'{class_name.upper()} Loss Curve')
plt.tight_layout()
plt.show()

# Model summaries
print("\n=== Model Architectures ===")
encoder.summary()
decoder.summary()
vae.summary()
```

```python
import os
import time
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import (Input, Dense, Lambda, Flatten, Reshape,
                                   Conv2D, Conv2DTranspose, Layer,
                                   MultiHeadAttention, LayerNormalization, Dropout,
                                   Concatenate)
from tensorflow.keras.models import Model
from tensorflow.keras import backend as K
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import matplotlib.pyplot as plt
from tensorflow.image import ssim, psnr
from tensorflow.keras.applications.inception_v3 import InceptionV3, preprocess_input
from scipy.linalg import sqrtm
from tensorflow.keras.callbacks import History
from tensorflow.keras.utils import plot_model

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Define image dimensions
img_shape = (128, 128, 3)
latent_dim = 100
max_images_per_class = 1000

# Performance tracking class
class PerformanceTracker:
    def __init__(self):
        self.metrics = {
            'training_time': [],
            'inference_time': [],
            'gpu_memory': [],
            'mse': [],
            'psnr': [],
            'ssim': [],
            'fid': []
        }

    def start_timer(self):
        self.start_time = time.time()

    def record_training_time(self):
        elapsed = time.time() - self.start_time
        self.metrics['training_time'].append(elapsed)
        return elapsed

    def record_inference_time(self, batch_size):
        elapsed = (time.time() - self.start_time) / batch_size
        self.metrics['inference_time'].append(elapsed)
        return elapsed

    def get_gpu_memory(self):
        try:
            from pynvml import nvmlInit, nvmlDeviceGetHandleByIndex, nvmlDeviceGetMemoryInfo
            nvmlInit()
            handle = nvmlDeviceGetHandleByIndex(0)
            info = nvmlDeviceGetMemoryInfo(handle)
            return info.used // 1024 // 1024  # Convert to MB
        except:
            return None

    def record_metrics(self, mse, psnr, ssim, fid):
        self.metrics['mse'].append(mse)
        self.metrics['psnr'].append(psnr)
        self.metrics['ssim'].append(ssim)
        self.metrics['fid'].append(fid)

    def print_summary(self, class_name):
        print(f"\n=== Performance Summary for {class_name.upper()} ===")
        print(f"Training time: {np.mean(self.metrics['training_time']):.2f} ± {np.std(self.metrics['training_time']):.2f} sec")
        print(f"Inference time: {np.mean(self.metrics['inference_time']) * 1000:.2f} ± {np.std(self.metrics['inference_time']) * 1000:.2f} ms per image")
        if any(self.metrics['gpu_memory']):
            print(f"GPU memory usage: {np.mean(self.metrics['gpu_memory']):.0f} ± {np.std(self.metrics['gpu_memory']):.0f} MB")
        print(f"Reconstruction quality:")
        print(f"- MSE: {np.mean(self.metrics['mse']):.4f}")
        print(f"- PSNR: {np.mean(self.metrics['psnr']):.2f} dB")
        print(f"- SSIM: {np.mean(self.metrics['ssim']):.4f}")
        print(f"- FID: {np.mean(self.metrics['fid']):.2f}")

# Function to Load Images
def load_images(directory, img_size):
    images = []
    for filename in os.listdir(directory):
        img_path = os.path.join(directory, filename)
        try:
            img = load_img(img_path, target_size=img_size)
            img = img_to_array(img) / 255.0
            images.append(img)
        except Exception as e:
            print(f"Error loading {img_path}: {e}")
    return np.array(images)

# Load datasets for all three classes
base_dir = '/content/drive/MyDrive/BRATS20/Pre_BraTS20_flair'
class_dirs = {
    'Tumor': os.path.join(base_dir, 'Tumor'),
    'NoTumor': os.path.join(base_dir, 'NoTumor')
}

# Check if directories exist and load images
x_train = {}
for class_name, data_dir in class_dirs.items():
    if not os.path.exists(data_dir):
        raise ValueError(f"Directory {data_dir} does not exist! Check the path.")
    x_train[class_name] = load_images(data_dir, (128, 128))
    print(f"Loaded {len(x_train[class_name])} images for class {class_name}")

# Custom Transformer Block Layer with Cross-Attention
class CrossAttentionTransformerBlock(Layer):
    def __init__(self, num_heads=8, key_dim=64, dropout_rate=0.1, **kwargs):
        super().__init__(**kwargs)
        self.num_heads = num_heads
        self.key_dim = key_dim
        self.dropout_rate = dropout_rate

        # Cross-attention layers
        self.cross_attn = MultiHeadAttention(num_heads=num_heads, key_dim=key_dim, dropout=dropout_rate)
        self.dropout1 = Dropout(dropout_rate)
        self.layernorm1 = LayerNormalization(epsilon=1e-6)

        # Feed-forward network
        self.ffn = tf.keras.Sequential([
            Dense(key_dim * 4, activation='relu'),
            Dense(key_dim),
            Dropout(dropout_rate)
        ])
        self.layernorm2 = LayerNormalization(epsilon=1e-6)

    def call(self, inputs):
        # Unpack inputs: latent_z is query, encoder_features are key/value
        latent_z, encoder_features = inputs

        # Cross-attention
        attn_output = self.cross_attn(
            query=latent_z,
            key=encoder_features,
            value=encoder_features
        )
        attn_output = self.dropout1(attn_output)
        out1 = self.layernorm1(latent_z + attn_output)

        # Feed-forward network
        ffn_output = self.ffn(out1)
        return self.layernorm2(out1 + ffn_output)

    def get_config(self):
        config = super().get_config()
        config.update({
            'num_heads': self.num_heads,
            'key_dim': self.key_dim,
            'dropout_rate': self.dropout_rate
        })
        return config

# Enhanced Encoder that returns intermediate features
def build_encoder():
    inputs = Input(shape=img_shape)

    # First conv block
    x = Conv2D(32, kernel_size=3, strides=2, padding='same', activation='relu')(inputs)

    # Second conv block (we'll use these features for cross-attention)
    encoder_features = Conv2D(64, kernel_size=3, strides=2, padding='same', activation='relu')(x)

    # Flatten and get latent distribution parameters
    x = Flatten()(encoder_features)
    x = Dense(256, activation='relu')(x)

    z_mean = Dense(latent_dim, name='z_mean')(x)
    z_log_var = Dense(latent_dim, name='z_log_var')(x)

    # Sampling function
    def sampling(args):
        z_mean, z_log_var = args
        batch = K.shape(z_mean)[0]
        dim = K.int_shape(z_mean)[1]
        epsilon = K.random_normal(shape=(batch, dim))
        return z_mean + K.exp(0.5 * z_log_var) * epsilon

    z = Lambda(sampling, output_shape=(latent_dim,), name='z')([z_mean, z_log_var])

    # Return both latent variables and intermediate features
    return Model(inputs, [z_mean, z_log_var, z, encoder_features], name='encoder')

# Enhanced Decoder with Cross-Attention Transformer
def build_decoder():
    # Two inputs: latent vector and encoder features
    latent_inputs = Input(shape=(latent_dim,))
    encoder_features_input = Input(shape=(32, 32, 64))  # Shape of encoder's intermediate features

    # Process latent vector
    x = Dense(16 * 16 * 64, activation='relu')(latent_inputs)
    x = Reshape((16, 16, 64))(x)

    # Prepare for cross-attention
    latent_flat = Reshape((16*16, 64))(x)  # Query (from latent)
    encoder_flat = Reshape((32*32, 64))(encoder_features_input)  # Key/Value (from encoder)

    # Apply cross-attention
    x = CrossAttentionTransformerBlock()([latent_flat, encoder_flat])
    x = Reshape((16, 16, 64))(x)

    # Upsampling
    x = Conv2DTranspose(64, kernel_size=3, strides=2, padding='same', activation='relu')(x)
    x = Conv2DTranspose(32, kernel_size=3, strides=2, padding='same', activation='relu')(x)
    outputs = Conv2DTranspose(3, kernel_size=3, strides=2, padding='same', activation='sigmoid')(x)

    return Model([latent_inputs, encoder_features_input], outputs, name='decoder_with_cat')

# Custom Loss Layer
class VAELossLayer(Layer):
    def __init__(self, **kwargs):
        super(VAELossLayer, self).__init__(**kwargs)

    def vae_loss(self, inputs, outputs, z_mean, z_log_var):
        reconstruction_loss = tf.keras.losses.binary_crossentropy(
            K.flatten(inputs), K.flatten(outputs))
        reconstruction_loss = tf.reduce_mean(reconstruction_loss) * np.prod(img_shape)

        kl_loss = -0.5 * tf.reduce_sum(
            1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var), axis=-1)
        return tf.reduce_mean(reconstruction_loss + kl_loss)

    def call(self, inputs):
        inputs, outputs, z_mean, z_log_var = inputs
        loss = self.vae_loss(inputs, outputs, z_mean, z_log_var)
        self.add_loss(loss)
        return outputs

def build_vae(encoder, decoder):
    inputs = Input(shape=img_shape)
    z_mean, z_log_var, z, encoder_features = encoder(inputs)
    outputs = decoder([z, encoder_features])
    outputs = VAELossLayer()([inputs, outputs, z_mean, z_log_var])
    return Model(inputs, outputs, name='vae_with_cat')

# Function to calculate evaluation metrics
def evaluate_model(encoder, decoder, vae, x_test):
    reconstructed_images = vae.predict(x_test)

    mse_scores = np.mean(np.square(x_test - reconstructed_images), axis=(1, 2, 3))
    psnr_scores = [psnr(x_test[i], reconstructed_images[i], max_val=1.0).numpy() for i in range(len(x_test))]
    ssim_scores = [ssim(x_test[i], reconstructed_images[i], max_val=1.0).numpy() for i in range(len(x_test))]

    avg_mse = np.mean(mse_scores)
    avg_psnr = np.mean(psnr_scores)
    avg_ssim = np.mean(ssim_scores)

    return avg_mse, avg_psnr, avg_ssim

# Load InceptionV3 model for FID calculation
fid_model = InceptionV3(include_top=False, pooling='avg', input_shape=(128, 128, 3))

# Function to compute FID Score
def calculate_fid(real_images, generated_images):
    # Preprocess images for InceptionV3
    real_images = preprocess_input(real_images)
    generated_images = preprocess_input(generated_images)

    # Compute feature representations
    act_real = fid_model.predict(real_images)
    act_generated = fid_model.predict(generated_images)

    # Calculate mean and covariance statistics
    mu_real, sigma_real = np.mean(act_real, axis=0), np.cov(act_real, rowvar=False)
    mu_generated, sigma_generated = np.mean(act_generated, axis=0), np.cov(act_generated, rowvar=False)

    # Compute squared difference of means
    mean_diff = np.sum((mu_real - mu_generated) ** 2)

    # Compute sqrt of product of covariances
    covmean, _ = sqrtm(sigma_real.dot(sigma_generated), disp=False)

    # Check for imaginary values
    if np.iscomplexobj(covmean):
        covmean = covmean.real

    # Compute FID score
    fid_score = mean_diff + np.trace(sigma_real + sigma_generated - 2 * covmean)
    return fid_score

# Train separate VAEs for each class
for class_name, images in x_train.items():
    tracker = PerformanceTracker()
    print(f"\n=== Training CAT-VAE for class: {class_name.upper()} ===")
    print(f"Training samples: {len(images)}")

    # Build models
    encoder = build_encoder()
    decoder = build_decoder()
    vae = build_vae(encoder, decoder)
    vae.compile(optimizer='adam')

    # Start training timer
    tracker.start_timer()

    # Train the model and store history
    history = vae.fit(images, epochs=500, batch_size=64, verbose=1)

    # Plot training history
    plt.figure(figsize=(10, 6))
    plt.plot(history.history['loss'], label='Training Loss')
    plt.title(f'Training Loss Curve for {class_name.upper()}')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    # Save the loss curve plot
    loss_curve_path = os.path.join(base_dir, f'loss_curve_{class_name}.png')
    plt.savefig(loss_curve_path)
    plt.close()
    print(f"Saved loss curve plot at: {loss_curve_path}")

    # Record training time
    training_time = tracker.record_training_time()
    print(f"Training completed in {training_time:.2f} seconds")

    # Record GPU memory
    gpu_mem = tracker.get_gpu_memory()
    if gpu_mem:
        tracker.metrics['gpu_memory'].append(gpu_mem)
        print(f"GPU memory used: {gpu_mem} MB")

    # Create a directory to save models for this class
    models_dir = os.path.join(base_dir, f'saved_models_VAE-CAT_Brain{class_name}')
    os.makedirs(models_dir, exist_ok=True)

    # Save models
    encoder.save(os.path.join(models_dir, 'encoder.h5'))
    decoder.save(os.path.join(models_dir, 'decoder.h5'))
    vae.save(os.path.join(models_dir, 'vae.h5'))

    # Save model architecture diagram
    architecture_path = os.path.join(models_dir, 'vae_architecture.png')
    try:
        plot_model(vae, to_file=architecture_path, show_shapes=True, show_layer_names=True)
        print(f"Saved model architecture diagram at: {architecture_path}")
    except Exception as e:
        print(f"Error saving model architecture diagram: {e}")

    print(f"Saved models for {class_name} in {models_dir}")

    # Create output directory for generated images
    output_dir = os.path.join(base_dir, f'generated_VAE-CAT_Brain{class_name}')
    os.makedirs(output_dir, exist_ok=True)

    # Calculate how many images to generate
    num_real_images = len(images)
    num_to_generate = max(0, max_images_per_class - num_real_images)

    if num_to_generate > 0:
        print(f"Generating {num_to_generate} images for class {class_name}")

        # Generate latent vectors
        latent_vectors = np.random.normal(size=(num_to_generate, latent_dim))

        # Need encoder features for generation - use average features from training set
        _, _, _, sample_encoder_features = encoder.predict(images[:min(100, len(images))])
        avg_encoder_features = np.mean(sample_encoder_features, axis=0, keepdims=True)
        repeated_features = np.repeat(avg_encoder_features, num_to_generate, axis=0)

        # Generate images
        generated_images = decoder.predict([latent_vectors, repeated_features])

        # Save generated images
        for i in range(num_to_generate):
            img_path = os.path.join(output_dir, f'generated_{i}.png')
            plt.imsave(img_path, generated_images[i])

    # Evaluate the model
    print(f"\nEvaluating model for class: {class_name}")
    x_test_sample = images[:min(10, len(images))]

    # Measure inference speed
    tracker.start_timer()
    _ = vae.predict(x_test_sample)
    inference_time = tracker.record_inference_time(len(x_test_sample))
    print(f"Inference speed: {inference_time*1000:.2f} ms per image")

    # Calculate quality metrics
    mse, psnr_score, ssim_score = evaluate_model(encoder, decoder, vae, x_test_sample)
    print(f"Reconstruction quality:")
    print(f"- MSE: {mse:.4f}")
    print(f"- PSNR: {psnr_score:.2f} dB")
    print(f"- SSIM: {ssim_score:.4f}")

    # Calculate FID score
    if len(x_test_sample) > 0:
        # Generate comparable number of images for FID
        latent_vectors = np.random.normal(size=(len(x_test_sample), latent_dim))
        _, _, _, sample_encoder_features = encoder.predict(x_test_sample)
        generated_for_fid = decoder.predict([latent_vectors, sample_encoder_features])
        fid_score = calculate_fid(x_test_sample, generated_for_fid)
        print(f"- FID: {fid_score:.2f}")
    else:
        fid_score = float('nan')

    # Record all metrics
    tracker.record_metrics(mse, psnr_score, ssim_score, fid_score)

    # Print summary
    tracker.print_summary(class_name)

# Create and save a comprehensive system diagram
try:
    # Create a temporary model that shows the full flow
    inputs = Input(shape=img_shape)
    z_mean, z_log_var, z, encoder_features = encoder(inputs)
    outputs = decoder([z, encoder_features])
    full_system = Model(inputs, outputs, name='full_vae_system')

    system_diagram_path = os.path.join(base_dir, 'full_system_architecture.png')
    plot_model(full_system, to_file=system_diagram_path, show_shapes=True, show_layer_names=True)
    print(f"\nSaved full system architecture diagram at: {system_diagram_path}")
except Exception as e:
    print(f"\nError creating full system diagram: {e}")

# Display all loss curves at the end
plt.figure(figsize=(15, 10))
for i, class_name in enumerate(x_train.keys()):
    loss_curve_path = os.path.join(base_dir, f'loss_curve_{class_name}.png')
    if os.path.exists(loss_curve_path):
        img = plt.imread(loss_curve_path)
        plt.subplot(2, 2, i+1)
        plt.imshow(img)
        plt.axis('off')
        plt.title(f'{class_name.upper()} Loss Curve')
plt.tight_layout()
plt.show()

# Model summaries
print("\n=== Model Architectures ===")
encoder.summary()
decoder.summary()
vae.summary()
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import load_model

# Define parameters
latent_dim = 100
num_images_to_generate = 1000
output_dir = '/content/Brain1'
os.makedirs(output_dir, exist_ok=True)

# Register the sampling function (must match exactly how it was registered during training)
def sampling(args):
    z_mean, z_log_var = args
    batch = tf.shape(z_mean)[0]
    dim = tf.shape(z_mean)[1]
    epsilon = tf.keras.backend.random_normal(shape=(batch, dim))
    return z_mean + tf.exp(0.5 * z_log_var) * epsilon

# Function to load models and generate images
def generate_images(class_name, base_dir):
    print(f"\n=== Generating images for class: {class_name.upper()} ===")

    # Path to saved models
    models_dir = os.path.join(base_dir, f'saved_models_VAE-CAT_Brain{class_name}')

    try:
        # Load models with custom objects
        custom_objects = {
            'sampling': sampling,
            'VAELossLayer': VAELossLayer,
            'CrossAttentionTransformerBlock': CrossAttentionTransformerBlock
        }

        encoder = load_model(
            os.path.join(models_dir, 'encoder.h5'),
            custom_objects=custom_objects,
            compile=False
        )

        decoder = load_model(
            os.path.join(models_dir, 'decoder.h5'),
            custom_objects=custom_objects,
            compile=False
        )

        print("Models loaded successfully")

        # Generate latent vectors
        latent_vectors = np.random.normal(size=(num_images_to_generate, latent_dim))

        # Create dummy features with correct shape (32, 32, 64)
        dummy_features = np.zeros((num_images_to_generate, 32, 32, 64))

        # Generate images
        generated_images = decoder.predict([latent_vectors, dummy_features])

        # Save images
        class_output_dir = os.path.join(output_dir, class_name)
        os.makedirs(class_output_dir, exist_ok=True)

        for i in range(num_images_to_generate):
            img_path = os.path.join(class_output_dir, f'generated_{class_name}_{i}.png')
            plt.imsave(img_path, generated_images[i])
            if i % 10 == 0:
                print(f"Saved image {i+1}/{num_images_to_generate}")

        print(f"Successfully generated and saved {num_images_to_generate} images to {class_output_dir}")

    except Exception as e:
        print(f"Error generating images for {class_name}: {str(e)}")

# You'll need to include these class definitions (same as in your training code)
class VAELossLayer(tf.keras.layers.Layer):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)

    def vae_loss(self, inputs, outputs, z_mean, z_log_var):
        reconstruction_loss = tf.keras.losses.binary_crossentropy(
            tf.keras.backend.flatten(inputs),
            tf.keras.backend.flatten(outputs))
        reconstruction_loss = tf.reduce_mean(reconstruction_loss) * np.prod((128, 128, 3))

        kl_loss = -0.5 * tf.reduce_sum(
            1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var), axis=-1)
        return tf.reduce_mean(reconstruction_loss + kl_loss)

    def call(self, inputs):
        inputs, outputs, z_mean, z_log_var = inputs
        loss = self.vae_loss(inputs, outputs, z_mean, z_log_var)
        self.add_loss(loss)
        return outputs

class CrossAttentionTransformerBlock(tf.keras.layers.Layer):
    def __init__(self, num_heads=8, key_dim=64, dropout_rate=0.1, **kwargs):
        super().__init__(**kwargs)
        self.num_heads = num_heads
        self.key_dim = key_dim
        self.dropout_rate = dropout_rate
        self.cross_attn = tf.keras.layers.MultiHeadAttention(
            num_heads=num_heads, key_dim=key_dim, dropout=dropout_rate)
        self.dropout1 = tf.keras.layers.Dropout(dropout_rate)
        self.layernorm1 = tf.keras.layers.LayerNormalization(epsilon=1e-6)
        self.ffn = tf.keras.Sequential([
            tf.keras.layers.Dense(key_dim * 4, activation='relu'),
            tf.keras.layers.Dense(key_dim),
            tf.keras.layers.Dropout(dropout_rate)
        ])
        self.layernorm2 = tf.keras.layers.LayerNormalization(epsilon=1e-6)

    def call(self, inputs):
        latent_z, encoder_features = inputs
        attn_output = self.cross_attn(
            query=latent_z,
            key=encoder_features,
            value=encoder_features
        )
        attn_output = self.dropout1(attn_output)
        out1 = self.layernorm1(latent_z + attn_output)
        ffn_output = self.ffn(out1)
        return self.layernorm2(out1 + ffn_output)

    def get_config(self):
        config = super().get_config()
        config.update({
            'num_heads': self.num_heads,
            'key_dim': self.key_dim,
            'dropout_rate': self.dropout_rate
        })
        return config

# Base directory where models are saved
base_dir = '/content/drive/MyDrive/BRATS20/Pre_BraTS20_flair'

# Generate images for each class
for class_name in ['Tumor', 'NoTumor']:
    generate_images(class_name, base_dir)

print("\nImage generation complete!")
```
